```python
import pandas as pd
```

```python
df = pd.read_csv('competition_53_season_315_all_matches.csv')
df
```

```python
import numpy as np

df.loc[df['pass_outcome'].replace('', np.nan).isna(), 'pass_outcome'] = 'Complete'
```

```python
df['pass_outcome'].value_counts()
```

```python
# Create fd with unique player and their team
fd = df[['player', 'team']].drop_duplicates(subset='player').reset_index(drop=True)

fd
```

```python
# Count Passes Attempted per player
pass_counts = df[df['type'] == 'Pass'].groupby('player').size()

# Count Carries Made per player
carry_counts = df[df['type'] == 'Carry'].groupby('player').size()

# Create fd with player, team, and counts
fd = df[['player', 'team']].drop_duplicates(subset='player').reset_index(drop=True)

# Map counts to fd (fill missing with 0)
fd['Passes Attempted'] = fd['player'].map(pass_counts).fillna(0).astype(int)
fd['Carries Made'] = fd['player'].map(carry_counts).fillna(0).astype(int)

fd
```

```python
# Count successful passes: type=Pass and pass_outcome='Complete'
successful_pass_counts = df[(df['type'] == 'Pass') & (df['pass_outcome'] == 'Complete')].groupby('player').size()

# Add to fd, fill missing with 0
fd['Successful passes Made'] = fd['player'].map(successful_pass_counts).fillna(0).astype(int)

# Calculate Pass Completion Rate (avoid division by zero)
fd['Pass Completion rate'] = fd.apply(
    lambda row: row['Successful passes Made'] / row['Passes Attempted'] if row['Passes Attempted'] > 0 else 0,
    axis=1
)

fd
```

```python
fd['Pass Completion rate'] = fd['Pass Completion rate'] * 100
fd
```

```python
fd['Pass Completion rate'] = fd.apply(
    lambda row: (row['Successful passes Made'] / row['Passes Attempted'] * 100) if row['Passes Attempted'] > 0 else 0,
    axis=1
).round(2)
fd
```

```python
# 1. Progressive Pass Attempted (progressive_pass = True)
progressive_pass_attempted = df[(df['type'] == 'Pass') & (df['progressive_pass'] == True)].groupby('player').size()

# 2. Successful Progressive Passes (pass_outcome = 'Complete' and progressive_pass = True)
successful_progressive_passes = df[
    (df['type'] == 'Pass') & 
    (df['pass_outcome'] == 'Complete') & 
    (df['progressive_pass'] == True)
].groupby('player').size()

# 3. Successful Progressive Passes Under Pressure (pass_outcome = 'Complete', progressive_pass = True, under_pressure = True)
successful_progressive_passes_under_pressure = df[
    (df['type'] == 'Pass') & 
    (df['pass_outcome'] == 'Complete') & 
    (df['progressive_pass'] == True) & 
    (df['under_pressure'] == True)
].groupby('player').size()

# Add these counts to fd, fill missing values with 0 and convert to int
fd['Progressive Pass Attempted'] = fd['player'].map(progressive_pass_attempted).fillna(0).astype(int)
fd['Successful Progressive Passes'] = fd['player'].map(successful_progressive_passes).fillna(0).astype(int)
fd['Successful Progressive Passes Under Pressure'] = fd['player'].map(successful_progressive_passes_under_pressure).fillna(0).astype(int)

# 4. Progressive Pass Completion rate (rounded to two decimals)
fd['Progressive Pass Completion rate'] = fd.apply(
    lambda row: (row['Successful Progressive Passes'] / row['Progressive Pass Attempted'] * 100) if row['Progressive Pass Attempted'] > 0 else 0,
    axis=1
).round(2)

fd
```

```python
# Progressive Ball Carries (type = Carry and progressive_carry = True)
progressive_ball_carries = df[
    (df['type'] == 'Carry') & 
    (df['progressive_carry'] == True)
].groupby('player').size()

# Progressive Ball Carries Under Pressure (type = Carry and progressive_carry = True and under_pressure = True)
progressive_ball_carries_under_pressure = df[
    (df['type'] == 'Carry') & 
    (df['progressive_carry'] == True) & 
    (df['under_pressure'] == True)
].groupby('player').size()

# Map counts to fd, fill missing with 0 and convert to int
fd['Progressive Ball Carries'] = fd['player'].map(progressive_ball_carries).fillna(0).astype(int)
fd['Progressive Ball Carries Under Pressure'] = fd['player'].map(progressive_ball_carries_under_pressure).fillna(0).astype(int)

fd
```

```python
# Line Breaking Passes Attempted
line_breaking_passes_attempted = df[
    (df['type'] == 'Pass') &
    (df['progressive_pass'] == True) &
    (df['line_breaking'] == True)
].groupby('player').size()

# Line Breaking Passes Completed
line_breaking_passes_completed = df[
    (df['type'] == 'Pass') &
    (df['progressive_pass'] == True) &
    (df['line_breaking'] == True) &
    (df['pass_outcome'] == 'Complete')
].groupby('player').size()

# Line Breaking Passes Completed Under Pressure
line_breaking_passes_completed_under_pressure = df[
    (df['type'] == 'Pass') &
    (df['progressive_pass'] == True) &
    (df['line_breaking'] == True) &
    (df['pass_outcome'] == 'Complete') &
    (df['under_pressure'] == True)
].groupby('player').size()

# Map counts to fd, fill missing with 0 and convert to int
fd['Line Breaking Passes Attempted'] = fd['player'].map(line_breaking_passes_attempted).fillna(0).astype(int)
fd['Line Breaking Passes Completed'] = fd['player'].map(line_breaking_passes_completed).fillna(0).astype(int)
fd['Line Breaking Passes Completed Under Pressure'] = fd['player'].map(line_breaking_passes_completed_under_pressure).fillna(0).astype(int)

# Calculate Line Breaking Pass Completion Rate
fd['Line Breaking Pass Completion Rate'] = (
    fd['Line Breaking Passes Completed'] / fd['Line Breaking Passes Attempted'].replace(0, pd.NA) * 100
).round(2).fillna(0)

fd
```
