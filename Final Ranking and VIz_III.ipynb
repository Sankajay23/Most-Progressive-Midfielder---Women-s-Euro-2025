```python
import pandas as pd
```

```python
df = pd.read_csv('WEURO_Player_Data - Sheet1.csv')
df
```

```python
countries_to_drop = [
    "Sweden",
    "England",
    "Norway",
    "Italy",
    "France",
    "Germany",
    "Spain",
    "Switzerland"
]

# Create a single regex pattern: Sweden|England|Norway|...
pattern = "|".join(countries_to_drop)

# Drop rows where team contains any of the country names
df = df[~df['team'].str.contains(pattern, case=False, na=False)]

df
```

```python
cols_to_keep = ['player', 'team', '90s', 'MP']
per90_cols = [col for col in df.columns if 'Per 90' in col]
# Combine and filter
df = df[cols_to_keep + per90_cols]
```

```python
df = df[df['90s'] > 2.0]
df
```

```python
df.columns.tolist()
```

```python
import numpy as np


df['ProgPasses_per_SuccessPass'] = (
    df['Successful Progressive Passes Per 90'] /
    df['Successful passes Made Per 90'].replace(0, np.nan)
)

df['ProgPassesUnderPressure_per_SuccessPass'] = (
    df['Successful Progressive Passes Under Pressure Per 90'] /
    df['Successful passes Made Per 90'].replace(0, np.nan)
)

df['LineBreaking_per_SuccessPass'] = (
    df['Line Breaking Passes Completed Per 90'] /
    df['Successful passes Made Per 90'].replace(0, np.nan)
)

df['LineBreakingUnderPressure_per_SuccessPass'] = (
    df['Line Breaking Passes Completed Under Pressure Per 90'] /
    df['Successful passes Made Per 90'].replace(0, np.nan)
)

# --- Ratios relative to Carries Made Per 90 ---
df['ProgCarries_per_Carry'] = (
    df['Progressive Ball Carries Per 90'] /
    df['Carries Made Per 90'].replace(0, np.nan)
)

df['ProgCarriesUnderPressure_per_Carry'] = (
    df['Progressive Ball Carries Under Pressure Per 90'] /
    df['Carries Made Per 90'].replace(0, np.nan)
)
```

```python
df
```

```python
cols_to_convert = [
    'ProgPasses_per_SuccessPass',
    'ProgPassesUnderPressure_per_SuccessPass',
    'LineBreaking_per_SuccessPass',
    'LineBreakingUnderPressure_per_SuccessPass',
    'ProgCarries_per_Carry',
    'ProgCarriesUnderPressure_per_Carry'
]

df[cols_to_convert] = df[cols_to_convert] * 100
df
```

```python
import pandas as pd

# All ratio columns
ratio_cols = [
    'ProgPasses_per_SuccessPass',
    'ProgPassesUnderPressure_per_SuccessPass',
    'LineBreaking_per_SuccessPass',
    'LineBreakingUnderPressure_per_SuccessPass',
    'ProgCarries_per_Carry',
    'ProgCarriesUnderPressure_per_Carry',
    'Pass Completion rate Per 90'
]

# 1️⃣ Round to single decimal
df[ratio_cols] = df[ratio_cols].round(1)

# 2️⃣ Compute percentile rank for each
for col in ratio_cols:
    rank_col = f'{col}_RankPct'
    df[rank_col] = df[col].rank(pct=True) * 100

# 3️⃣ Define which columns to include in total percentile (exclude "Under Pressure")
included_for_total = [
    'ProgPasses_per_SuccessPass_RankPct',
    'LineBreaking_per_SuccessPass_RankPct',
    'ProgCarries_per_Carry_RankPct',
    'Pass Completion rate Per 90_RankPct'
]

# 4️⃣ Compute total percentile as **sum** of selected ranks
df['Total_Rank_Percentile'] = df[included_for_total].sum(axis=1).round(1)

# 5️⃣ Get top 20 players with highest total percentile
top201 = df.sort_values('Total_Rank_Percentile', ascending=False).head(20)
```

```python
top201
```

```python
top20_prog_pass = (
    df.sort_values('ProgPasses_per_SuccessPass_RankPct', ascending=False)
      .head(20)
)

top20_line_break = (
    df.sort_values('LineBreaking_per_SuccessPass_RankPct', ascending=False)
      .head(20)
)

top20_prog_carry = (
    df.sort_values('ProgCarries_per_Carry_RankPct', ascending=False)
      .head(20)
)

# Optional: keep only the most relevant columns for display
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Sort by total value
top201_sorted = top201.sort_values(by="Total_Rank_Percentile", ascending=False).reset_index(drop=True)

# Columns to stack
stack_cols = [
    'ProgPasses_per_SuccessPass_RankPct',
    'LineBreaking_per_SuccessPass_RankPct',
    'ProgCarries_per_Carry_RankPct',
    'Pass Completion rate Per 90_RankPct'
]

# Player names
players = top201_sorted['player'].tolist()

# Values for stacking
stack_values = [top201_sorted[col].tolist() for col in stack_cols]

# Color map (4 colours evenly spaced)
colors = cm.GnBu([i / len(stack_cols) for i in range(len(stack_cols))])

plt.figure(figsize=(12, 6))

bottom = [0] * len(top201_sorted)

# Plot stacked bars
for i, col in enumerate(stack_cols):
    plt.bar(
        players,
        stack_values[i],
        bottom=bottom,
        label=col.replace('_', ' '),
        color=colors[i]
    )
    bottom = [a + b for a, b in zip(bottom, stack_values[i])]

plt.title("Stacked Rank Percentile Contributions – Top 20 Players")
plt.ylabel("Rank Percentile Contribution")
plt.xticks(rotation=75)

# Smaller legend
plt.legend(prop={'size': 8})

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="white")

# 1. Progressive Passes per Success Pass
top20_prog_pass_sorted = top20_prog_pass.sort_values('ProgPasses_per_SuccessPass', ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(
    x='ProgPasses_per_SuccessPass', 
    y='player', 
    data=top20_prog_pass_sorted, 
    palette="GnBu"
)
plt.title('Top 20 Players - Progressive Passes per Success Pass')
plt.xlabel('Progressive Passes per Success Pass')
plt.ylabel('Player')
plt.show()

# 2. Line Breaking per Success Pass
top20_line_break_sorted = top20_line_break.sort_values('LineBreaking_per_SuccessPass', ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(
    x='LineBreaking_per_SuccessPass', 
    y='player', 
    data=top20_line_break_sorted, 
    palette="GnBu"
)
plt.title('Top 20 Players - Line Breaking per Success Pass')
plt.xlabel('Line Breaking per Success Pass')
plt.ylabel('Player')
plt.show()

# 3. Progressive Carries per Carry
top20_prog_carry_sorted = top20_prog_carry.sort_values('ProgCarries_per_Carry', ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(
    x='ProgCarries_per_Carry', 
    y='player', 
    data=top20_prog_carry_sorted, 
    palette="GnBu"
)
plt.title('Top 20 Players - Progressive Carries per Carry')
plt.xlabel('Progressive Carries per Carry')
plt.ylabel('Player')
plt.show()
```
